```python
import spacy
import networkx as nx
from itertools import combinations
```

```python
nlp = spacy.load("en_core_web_sm")
```

```python
def textrank_keywords(text, top_n=10):

    doc = nlp(text)

    words = [token.text.lower() for token in doc
             if token.pos_ in ["NOUN", "PROPN", "ADJ"]
             and not token.is_stop]

    graph = nx.Graph()

    for pair in combinations(words, 2):
        if graph.has_edge(pair[0], pair[1]):
            graph[pair[0]][pair[1]]['weight'] += 1
        else:
            graph.add_edge(pair[0], pair[1], weight=1)

    scores = nx.pagerank(graph)

    ranked = sorted(scores.items(), key=lambda x: x[1], reverse=True)

    return ranked[:top_n]
```

```python
text = """
Artificial Intelligence is transforming industries by enabling machines to
learn from data and make intelligent decisions. Machine learning and deep
learning techniques are widely used in healthcare, finance, and transportation.
"""
```

```python
keywords = textrank_keywords(text, top_n=10)

for word, score in keywords:
    print(word, score)
```
